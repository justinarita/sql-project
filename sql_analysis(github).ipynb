### PURPOSE
The purpose of this query is to find the discrepancies between Q1 iPhone sales and the other quarters. 

```python
%%sql
SELECT 'Quarter 1 Avg Sales' AS 'Fiscal Quarter Average', AVG(`Unit Sales (in millions)`) AS 'Average Quarterly Sales (in millions)', MAX(`Unit Sales (in millions)`) AS 'Most Ever Sales in a Quarter (in millions)'
FROM iPhoneUnitSales 
WHERE `Fiscal Quarter` LIKE 'Q1%'
UNION
SELECT 'Quarter 2,3,4 Avg Sales', AVG(`Unit Sales (in millions)`), max(`Unit Sales (in millions)`)
FROM iPhoneUnitSales 
WHERE `Fiscal Quarter` NOT LIKE 'Q1%';
```

### WHAT I DISCOVERED
Through the sql UNION, I was able to compare these two categories. The results displayed the strength of Q1 iPhone sales. The average for Q1 was around 16 million dollars higher than the average of the other three. I also saw signigicant differences when comparinf the highest selling quarter for each of the two categories. I showed this through using the average and max functions.

## Using SQL to answer the Apple and Samsung sales questions

### *Question*: Compare the trend and growth between Apple and Samsung by quarter.
#### *Business Justification*: We want to get a sense of any seasonal strength between the two companies. We are doing this by finding the amount of "above average quarters" for each company. If there is significant results, this could lead movement towards Apple capitalizing on a possible weak Samsung quarter 
#### *SQL Features Used*: VIEW, Aggregate (AVG), and Subqueries

```python
%%sql
CREATE OR REPLACE VIEW iphone_sales_vs_average AS
	SELECT
		`Fiscal Quarter`,
		`Unit Sales (in millions)`,
		ROUND((
			SELECT AVG(`Unit Sales (in millions)`)
			FROM iPhoneUnitSales
		), 3) AS avg_iphone_unit_sales,
		ROUND(`Unit Sales (in millions)` - 
		(
			SELECT AVG(`Unit Sales (in millions)`)
			FROM iPhoneUnitSales
		), 3) AS amt_above_avg
	FROM iPhoneUnitSales;
```

```python
%%sql
SELECT *
FROM iphone_sales_vs_average
WHERE `Unit Sales (in millions)` > (SELECT AVG(`Unit Sales (in millions)`)
									FROM iPhoneUnitSales
									);
```

```python
%%sql
CREATE OR REPLACE VIEW samsung_sales_vs_average AS 
	SELECT
		`Fiscal Quarter`,
		`Unit Sales (in millions)`,
		ROUND((
			SELECT AVG(`Unit Sales (in millions)`)
			FROM SamsungUnitSales
		), 3) AS avg_samsung_unit_sales,
		ROUND(`Unit Sales (in millions)` - 
		(
			SELECT AVG(`Unit Sales (in millions)`)
			FROM SamsungUnitSales
		), 3) AS amt_above_avg
FROM SamsungUnitSales;
```

```python
%%sql
SELECT *
FROM samsung_sales_vs_average
WHERE `Unit Sales (in millions)` > (SELECT AVG(`Unit Sales (in millions)`)
									FROM SamsungUnitSales
									);
```

### Recommendation
This comparison was impressive on Samsung's part as their, especially because there average is higher and they had more above average quarters than Apple did. But what is also did was show that their strength in sales doesn't come up in a specific quarter. As Apple, I would take this and continue to capitalize on a strong Q1 where customers know when to expect a new phone every year. Both companies are also able to use this to see what they did well and what they could do better on quarters where they are either well above or below their average quarterly sales.

### *Question*: Compare the accumulated growth between Apple and Samsung alongside the market share each quarter.
#### *Business Justification*: Here we are able to see a running total of sales as well as the worldwide market share percent of the company in chronological order.  Helps us to see side by side growth as well as worldwide reach in terms of the smartphone market.
#### *SQL Features Used*: ROW_NUM, JOIN, WINDOW

```python
%%sql
SELECT
	ROW_NUMBER() OVER() AS row_num,
	iPhoneUnitSales.`Fiscal Quarter`,
	ROUND(SUM(iPhoneUnitSales.`Unit Sales (in millions)`) OVER(ORDER BY iPhoneUnitSales.`Unit Sales (in millions)`), 2) AS 'iphone sales running total', 
	ROUND(SUM(SamsungUnitSales.`Unit Sales (in millions)`) OVER(ORDER BY SamsungUnitSales.`Unit Sales (in millions)`), 2) AS 'samsung sales running total', 
	GlobalMarketShare.Apple AS 'Apple world market share',
	GlobalMarketShare.Samsung AS 'Samsung world market share'
FROM iPhoneUnitSales
JOIN SamsungUnitSales
	ON iPhoneUnitSales.`Fiscal Quarter` = SamsungUnitSales.`Fiscal Quarter`
JOIN GlobalMarketShare
	ON SamsungUnitSales.`Fiscal Quarter` = GlobalMarketShare.`Fiscal Quarter`
ORDER BY row_num;
```

### Recommendation
We are able to see pretty crazy growth in both of these companies. The data is interesting as Apple had more of the global market share up until 2011, which is when Samsung saw an increase in Sales. Samsung hasn't looked back in the sense of leading the global smartphone market share. Managers of each company must look at these periods where samsung increased a lot, while apple didn't and vice versa as these points go to show that one company did something that the worldwide market liked over the other.

### *Question*: Compare worldwide smartphone sales for Apple and Samsung by showing sales growth in a year over year format for both Samsung and Apple.
#### *Business Justification*: Based on how we know Apple's strongest is Q1, we want to capitalize on this and continue to dig. Here we are able to see quarterly unit sales, previous quarter sales, year over year growth as well as the average sales for that quarter, historically. We are also able to see the way Q1 has grown by implementing a growth target of 10% year over year Q1 growth where it outputs either a target miss or beat.
#### *SQL Features Used*: LAG, ROUND, SUM, WINDOW, SUBQUERY, CASE, GROUP BY

```python
%%sql
SELECT
	`Fiscal Quarter`,
	`Unit Sales (in millions)` AS current_quarter_unit_sales,
	LAG(ROUND(SUM(`Unit Sales (in millions)`), 2)) OVER(ORDER BY `Fiscal Quarter`) AS previous_quarter_unit_sales,
	(SUM(`Unit Sales (in millions)`)
	- LAG(SUM(`Unit Sales (in millions)`),1) OVER(ORDER BY `Fiscal Quarter`))
	/ LAG(SUM(`Unit Sales (in millions)`),1) OVER(ORDER BY `Fiscal Quarter`)
	* 100 AS sales_growth_percent,
		CASE
			WHEN (SUM(`Unit Sales (in millions)`)
	- LAG(SUM(`Unit Sales (in millions)`),1) OVER(ORDER BY `Fiscal Quarter`))
	/ LAG(SUM(`Unit Sales (in millions)`),1) OVER(ORDER BY `Fiscal Quarter`)
	* 100 > 10 THEN 'Beat'
			ELSE 'Miss'
		END AS growth_target
FROM iPhoneUnitSales
GROUP BY `Fiscal Quarter`
HAVING `Fiscal Quarter` LIKE 'Q1%';
```

### Recommendation
A lot of growth is displayed in the beginning years, obviously as consumers rushed to buy iPhones. But something that stands out is the slow growth over the years 2016-2018. As a manager, I would immediately look into these years where the target has missed. You could look at the consumer market, as well as look internally towards possible features of the new iPhone. Overall, it gives direction to look at  what could have possibly halted the sales growth in these years/quarters.

### *Question*: What regions/countries could Apple potentially grow in?
#### *Business Justification*: Using google trends API, we will use SQL to find low/medium interest regions of the world that may be interested potentially in owning more iPhones. I found the average interest and made the high, medium, low metrics off of that. A company can target these areas with low exposure and interest and possibly sell to these places to increase sales. You can also use the view to find regions in the world with high interest, as well.
#### *SQL Features Used*: VIEW, CASE

```python
%%sql
CREATE OR REPLACE VIEW regional_iphone_interest AS
	SELECT `Region (Country)`, iPhone,
		CASE
			WHEN iPhone > 60 THEN 'High'
			WHEN iPhone BETWEEN 20 AND 60 THEN 'Medium'
			WHEN iPhone < 20 THEN 'Low'
		END AS iPhone_Interest 
	FROM InterestbyRegion;
```

```python
%%sql
SELECT *
FROM regional_iphone_interest
WHERE iphone_interest = 'Low' OR iphone_interest = 'Medium'
ORDER BY iphone;
```

As a manager and strategist, you can look at this view in a bunch of different ways. I presented it by opening up the query to low and medium interest countries. I would look at every one of these countries with a lens of potential and basically begin to dive in by researching why these countries are not interested in iphones. As a company like Apple, ask why are these companies not interested and what can we do to build interest?
